```python
import os
import pandas as pd
```

```python
cleaned_df = pd.read_csv("cleanedmedicaredata.csv")
cleaned_df.head()
```

```python
grouped_df = cleaned_df.groupby('state').agg(
    total_discharges=('discharges', 'sum'),
    avg_oop=('avg_oop', 'mean'),
    avg_insured=('avg_medicare_payment', 'mean')
).reset_index()

grouped_df.head()
```

```python
grouped_df.to_csv('stats_table.csv', index=False)
```
